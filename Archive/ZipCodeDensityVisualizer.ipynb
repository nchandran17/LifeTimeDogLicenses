```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:DB_PASSWORD@localhost:5432/SQL_Efficiency_Lab')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_query(file_name):
    with open(file_name, 'r') as f:
        return f.read()


query_string = load_query('../SQL_Queries/zipcode_density.sql')

df_zipcodes = pd.read_sql_query(query_string, engine)
plt.figure(figsize=(200, 100))
sns.set_theme(style="whitegrid")

sns.barplot(data=df_zipcodes, x = 'owner_zip', y = 'num_licenses_in_zip', hue='owner_zip', palette = 'plasma', legend=False, width=0.5)
plt.tick_params(axis='x', which='major', pad=15, labelsize=14)
plt.xticks(rotation=90, ha='right', fontsize=150)
plt.yticks(fontsize=150)
plt.title('Lifetime License ZipCode Density', fontsize=350, pad=40)
plt.xlabel('Zipcode', fontsize=200)
plt.ylabel('Number of Lifetime Licenses', fontsize=200)

# 5. Show it!
plt.show()
```
