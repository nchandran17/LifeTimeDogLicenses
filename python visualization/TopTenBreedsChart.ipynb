```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:DB_PASSWORD@localhost:5432/SQL_Efficiency_Lab')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_query(file_name):
    with open(file_name, 'r') as f:
        return f.read()
    
```

```python
query_string = load_query('../SQL_Queries/Top_Ten_Breeds.sql')

df_top_breeds = pd.read_sql_query(query_string, engine)
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# 3. Create the Bar Chart
sns.barplot(data=df_top_breeds, x='breed', y='num_of_breed', hue='breed', palette='viridis', legend=False)

# 4. Add the 'Executive' touches
plt.title('Top 10 Dog Breeds in Allegheny County', fontsize=16, pad=20)
plt.xlabel('Number of Registered Dogs', fontsize=12)
plt.ylabel('Breed', fontsize=12)

plt.savefig('top_10_breeds.png', dpi=300, bbox_inches='tight')
# 5. Show it!
plt.show()
```
