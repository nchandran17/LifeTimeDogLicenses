```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:DB_PASSWORD@localhost:5432/SQL_Efficiency_Lab')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_query(file_name):
    with open(file_name, 'r') as f:
        return f.read()
```

```python
query_string = load_query('../SQL_Queries/top_ten_colors.sql')
df_top_colors = pd.read_sql_query(query_string, engine)
plt.figure(figsize=(12, 6))
sns.set_theme(style="dark")

sns.barplot(x='color', y='num_of_color', data=df_top_colors, palette='viridis')
plt.title('Top 10 Colors in the Dataset')
plt.xlabel('Color', fontsize=12)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

plt.savefig('top_10_colors.png', dpi=300, bbox_inches='tight')
plt.show()
```
