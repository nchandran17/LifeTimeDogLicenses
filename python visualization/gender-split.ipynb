```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:DB_PASSWORD@localhost:5432/SQL_Efficiency_Lab')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import text
def load_query(fileName):
    with open(fileName, 'r') as file:
        query = file.read()
    return query


query_string = load_query('../SQL_Queries/num_of_each_gender.sql')
df_gender_split = pd.read_sql_query(text(query_string), engine)
plt.figure(figsize=(10, 6))
sns.set_theme(style = 'dark')
sns.barplot(x='gender', y='num_gender', data=df_gender_split, palette='rocket')
plt.title('Gender Split of Registered Dogs In Allegheny County', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Registered Dogs', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)\

plt.savefig('gender_split.png', dpi=300, bbox_inches='tight')
plt.show()
```
